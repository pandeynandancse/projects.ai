```

import urllib.request as urlreq
import urllib.parse
import urllib.error
import requests
from bs4 import BeautifulSoup
import ssl
import re
import  bs4
import json
import pandas as pd
#https://automatetheboringstuff.com/chapter11/
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
url = input("Enter the website page link you want to crawl :")
res = requests.get(url)

aa = bs4.BeautifulSoup(res.text)
selected = aa.select('script')
for select in selected:
  if select.get('src'):
    continue
  text = select.getText()
  if re.search("\t\t\t",text):
    #print('foud')
    break

splitted =  text.split("\t\t\t\tget_matrix(")
#print(len(splitted))


#for split in splitted:
#  print(split)
#  print("\n\n-------------------------------------------")
#print(splitted[2])
for split in splitted:
  if(re.search("^http",split[1:-4])):
    urlfind = split
    break
  else:
    continue

response = requests.get(urlfind[1:-4])
responded = response.json()


columns = []
for eachdict in responded['field']:
  columns.append(eachdict['name'])


df2 = pd.DataFrame(responded['data'],columns = columns)
df2
#print(urlfind[1:-4])

```
